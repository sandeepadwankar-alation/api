##Create a BI Server

This name configuration changes names "Report" to "View", "DataSource" to "BiSource" and "Connection" to "SQLSource". You can change to names to that applicable to the BI source

```
bi_server_details = [
  {
    "uri": "www.bi.com",
    "title": "Enterprise BI - 2020-1",
    "description": "Enterprise BI - 2020",
    "name_configuration": {
      "bi_report": "View",
      "bi_folder": "Folder",
      "bi_datasource": "BISource",
      "bi_connection": "SQLSource"
    }
  }
]
# ['bi_report', 'bi_datasource', 'bi_folder', 'bi_connection']
print(bi_server_details)
```

```
response = requests.post(bi_server_url, data=json.dumps(bi_server_details), headers=headers)
print(bi_server_url)
print(response.text)
print(response)
print(response.headers)
```

```
res = json.loads(response.text)
ids = res.get("Server IDs")
bi_server_id = ids[0]
bi_server_id
```

```
bi_server_api_url = (base_url + 'integration/v2/bi/server/' + str(bi_server_id))
bi_server_api_url
```
